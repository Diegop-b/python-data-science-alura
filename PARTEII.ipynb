```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
df = pd.read_json(url)


df_flat = pd.json_normalize(df.to_dict(orient="records"))
df_flat["Churn"] = df_flat["Churn"].astype(str).str.strip().str.lower()
df_flat["Churn"] = df_flat["Churn"].replace({
    "yes": "Sí",
    "no": "No",
    "nan": None,
    "": None
})

df_flat["account.Charges.Monthly"] = pd.to_numeric(df_flat["account.Charges.Monthly"], errors="coerce")
df_flat["Cuentas_Diarias"] = df_flat["account.Charges.Monthly"] / 30
print(df_flat["Cuentas_Diarias"].describe())

print("Distribución de Churn (clientes que se fueron o no):")
print(df_flat["Churn"].value_counts(normalize=True) * 100)

sns.countplot(data=df_flat, x="Churn", palette="pastel")
plt.title("¿Cuántos clientes abandonaron la empresa?")
plt.xlabel("Churn")
plt.ylabel("Cantidad de clientes")
plt.show()

variables_numericas = [
    "customer.tenure",
    "account.Charges.Monthly",
    "account.Charges.Total"

]

for var in variables_numericas:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df_flat, x="Churn", y=var, palette="coolwarm")
    plt.title(f"{var} según si el cliente se fue o no")
    plt.xlabel("Churn")
    plt.ylabel(var)
    plt.tight_layout()
    plt.show()

variables_categoricas = [
    "customer.gender",
    "customer.Partner",
    "customer.Dependents",
    "account.Contract",
    "account.PaymentMethod",
    "internet.InternetService",
    "phone.PhoneService"
]

for var in variables_categoricas:
    plt.figure(figsize=(7, 4))
    sns.countplot(data=df_flat, x=var, hue="Churn", palette="Set2")
    plt.title(f"{var} según si el cliente se fue o no")
    plt.xlabel(var)
    plt.ylabel("Cantidad")
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.histplot(df_flat["Cuentas_Diarias"], bins=40, kde=True, color="skyblue")
    plt.title("Distribución de las Cuentas Diarias")
    plt.xlabel("Cuentas Diarias ($)")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df_flat, x="Churn", y="Cuentas_Diarias", palette="coolwarm")
    plt.title("Cuentas Diarias según si el cliente se fue o no")
    plt.xlabel("Churn")
    plt.ylabel("Cuentas Diarias ($)")
    plt.tight_layout()
    plt.show()

variables_numericas = [
    "customer.tenure",
    "account.Charges.Monthly",
    "account.Charges.Total",
    "Cuentas_Diarias"
]

for col in variables_numericas:
    df_flat[col] = pd.to_numeric(df_flat[col], errors="coerce")

print("----- Estadísticas generales -----")
print(df_flat[variables_numericas].describe().T)

print("\n----- Medianas -----")
print(df_flat[variables_numericas].median())

print("\n----- Valores únicos por variable -----")
for var in variables_numericas:
    print(f"{var}: {df_flat[var].nunique()} valores únicos")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

df = pd.read_csv("TelecomX_clean.csv")

df.drop("customerID", axis=1, inplace=True)

df["Churn"] = df["Churn"].map({"Sí": 1, "No": 0})

df_encoded = pd.get_dummies(df, drop_first=True)

print("Valores nulos en 'Churn':", df_encoded["Churn"].isnull().sum())
print("Valores nulos totales:\n", df_encoded.isnull().sum().sort_values(ascending=False).head(5))


df_encoded.dropna(inplace=True)

X = df_encoded.drop("Churn", axis=1)
y = df_encoded["Churn"]

print("\nDistribución de clases:")
print(y.value_counts(normalize=True) * 100)

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("\nDespués de aplicar SMOTE:")
print(y_resampled.value_counts(normalize=True) * 100)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print(f"\nTamaño de entrenamiento: {X_train.shape}")
print(f"Tamaño de prueba: {X_test.shape}")
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)

print("=== Modelo 1: Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print("Matriz de Confusión:")
sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt='d', cmap='Blues')
plt.title("Confusión - Regresión Logística")
plt.show()
print(classification_report(y_test, y_pred_log))

X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)
y_pred_rf = rf_model.predict(X_test_rf)

print("=== Modelo 2: Random Forest ===")
print("Accuracy:", accuracy_score(y_test_rf, y_pred_rf))
print("Matriz de Confusión:")
sns.heatmap(confusion_matrix(y_test_rf, y_pred_rf), annot=True, fmt='d', cmap='Greens')
plt.title("Confusión - Random Forest")
plt.show()
print(classification_report(y_test_rf, y_pred_rf))
```
